```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('house_price_train.csv')
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df.head(6)
```

```python
df.isnull().sum()
```

```python
null_var=df.isnull().sum()/df.shape[0] *100
null_var
```

```python
drop_columns=null_var[null_var>20].keys()
drop_columns
```

```python
df=df.drop(columns=drop_columns)
```

```python
df.shape
```

```python
num_df=df.select_dtypes(include=['int64','float64'])
num_df.head()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(num_df.isnull())
```

```python
num_df.isnull().sum()
```

```python
missing_num_var=[var for var in num_df.columns if num_df[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(num_df[var],bins=20,rug_kws={'linewidth':5,'color':'#DC143C'})
```

```python
num_df2=num_df.fillna(num_df.mean())
```

```python
num_df2.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i ,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(num_df[var],bins=20,kde_kws={'linewidth':8,'color':'red'},label="Original")
    sns.distplot(num_df2[var],bins=20,kde_kws={'linewidth':5,'color':'green'},label="after filling mean value")
    plt.legend()
```

```python
num_df3=num_df.fillna(num_df.median())
plt.figure(figsize=(10,10))
sns.set()
for i ,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(num_df[var],bins=20,kde_kws={'linewidth':8,'color':'red'},label="Original")
    sns.distplot(num_df3[var],bins=20,kde_kws={'linewidth':5,'color':'green'},label="after filling median value")
    plt.legend()
```

```python
num_df4=num_df.fillna(num_df.mode())
plt.figure(figsize=(10,10))
sns.set()
for i ,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(num_df[var],bins=20,kde_kws={'linewidth':8,'color':'red'},label="Original")
    sns.distplot(num_df4[var],bins=20,kde_kws={'linewidth':5,'color':'green'},label="after filling mode value")
    plt.legend()
```
